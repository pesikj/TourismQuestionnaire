```python
import pandas as pd
from sklearn import tree
import graphviz
```

```python
df = pd.read_excel("../data/Spokojenost-a-profil-návštěvníka.xlsx")
df.head()

categorial_column_list = ["Kolikrát jste již v minulosti navštívil/a turistickou oblast Karlovarský kraj?", "Jaký je hlavní důvod Vaší návštěvy této oblasti?"]

X = df[categorial_column_list] = df[categorial_column_list].apply(lambda x: pd.factorize(x)[0])
y = pd.cut(df["Průměrná spokojenost"], bins=[0.9, 1.5, 2, 5], labels=["1", "2", "3"])
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
tree.plot_tree(clf)
dot_data = tree.export_graphviz(clf, out_file=None)
graph = graphviz.Source(dot_data)
graph.render("export")
```
